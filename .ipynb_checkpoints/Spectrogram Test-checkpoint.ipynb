```python
df_temp = {}
# 3,4,15,16,18,21,26
seizure_mag = np.zeros((129,0))
normal_mag = np.zeros((129,0))
for key, df_temp in zip(df.keys(), df.values()):
    temp = df_temp.as_matrix()
    x = np.mean(temp[:,3:], axis=1)
    seizure = temp[:,1]
    seizure_time = temp[:,0]
    seizure_start = seizure_time[np.argmax(seizure==1)]
    seizure_end = seizure_time[::-1][np.argmax(seizure[::-1]==1)]

    fs = 256.0
    f, t, Sxx = signal.spectrogram(x, fs, window=('tukey', 0.25))
    Sxx = 20*np.log10(Sxx)
    
    if '03' in key or '04' in key or '15' in key:
        plt.pcolormesh(t, f, Sxx)
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')
        plt.colorbar()
        plt.title('Spectrogram (dB)')
        plt.show()

        plt.figure()
        plt.plot(temp[:,0], seizure)
        plt.show()

    normal_mag = np.hstack([normal_mag, Sxx[:, (t < seizure_start)]])
    seizure_mag = np.hstack([seizure_mag, Sxx[:, (t >= seizure_start) & (t <= seizure_end)]])
    normal_mag = np.hstack([normal_mag, Sxx[:, (t > seizure_end)]])
plt.figure()
freq = np.arange(129)
plt.plot(freq, np.mean(normal_mag, axis=1),label='Normal')
seizure_mag[seizure_mag == -np.inf] = np.amin(seizure_mag[seizure_mag > -np.inf])
plt.plot(freq, np.mean(seizure_mag, axis=1), label='Seizure')
plt.xlabel('Frequenzy [Hz]')
plt.ylabel('Magnitude [dB]')
plt.legend()
plt.show()
```
